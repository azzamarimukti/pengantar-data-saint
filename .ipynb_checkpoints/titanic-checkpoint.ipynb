```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("titanic.csv")

# Cek total Mising Value
print("=" * 40)
print("=== Total Missing Value Perkolom ===")
print(df.isnull().sum())

# persentase missing value perkolom
print("\n")
print("=" * 40)
print("=== Persentase Missing Value Perkolom ===")
persen = ((df.isnull().sum() / len(df)*100).round(2))
print(persen)

# listwise Data Titanic
print("\n")
print("=" * 40)
print("=== Data Member Titanic ===")
df_listwise =df.dropna()
print(df_listwise)

# pairwise
# df_pairwise = df[['age']].dropna().corr()
# print(df_pairwise)

print("\n")
print("=" * 40)
print("=== rata Rata Umur ===")
rata2umur = np.mean(df["age"])
maksumur = np.max(df["age"])
minimumur = np.min(df["age"])
jarak_umur = np.std(df['age'])
print("Rata Rata umur: ", rata2umur)
print("umur yang paling tua: ", maksumur)
print("umur yang paling muda: ", minimumur)
print("Jarak/Perbedaan Umur: ", jarak_umur)

print("\n")
print("=" * 40)
print("== Calon di atas rata-rata usia ==")
# Filter data dengan usia di atas rata-rata
rata_usia = df['age'].mean()
calon_diatas_rata = df[df['age'] > rata_usia][['name', 'age']].head(6)
print(calon_diatas_rata.to_string(index=False))

print("\n")
print("=" * 40)
print("== Urutan dari Tertua ke Termuda ==")
# Urutkan data berdasarkan usia (descending)
urutan_usia = df[['name', 'age']].sort_values('age', ascending=False).head(10)
print(urutan_usia.to_string(index=False))

print("\n")
print("=" * 40)
print("== Data dengan Kategori Usia ==")
# Buat kategori usia
def kategori_usia(usia):
    if pd.isna(usia):
        return 'Tidak Diketahui'
    elif usia < 18:
        return 'Anak-anak'
    elif usia < 35:
        return 'Dewasa Muda'
    elif usia < 50:
        return 'Dewasa'
    else:
        return 'Lansia'

df['kategori_usia'] = df['age'].apply(kategori_usia)
data_kategori = df[['name', 'age', 'kategori_usia']].head(10)
print(data_kategori.to_string(index=False))

print("\n")
print("=" * 40)
print("== Jumlah Penumpang per Kategori Usia ==")
jumlah_per_kategori = df['kategori_usia'].value_counts()
print(jumlah_per_kategori)

print("\n")
print("=" * 40)
print("== Statistik Lanjutan ==")
print(f"Standar Deviasi Usia : {df['age'].std():.2f}")
print(f"Variansi Usia : {df['age'].var():.2f}")


print("\n")
print("=" * 40)
print("== Penumpang dengan Usia Tertua ==")
# Cari penumpang tertua
tertua = df.loc[df['age'].idxmax()]
print(f"Nama : {tertua['name']}")
print(f"Usia : {tertua['age']} tahun")

#bloxplot
kolom = ["age"]
sns.boxplot(data=df[kolom])
plt.title("Bloxplot Age")
plt.show()

```
