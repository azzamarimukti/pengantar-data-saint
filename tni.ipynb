```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("tni.csv")

print("== Data Calon TNI ==")
print(df.to_string(index=False))

print("\n" + "="*50)
print("== Statistik Dasar Tinggi Badan ==")
statistik = df['tinggi'].describe()
print(statistik)

print(f"\nRata-rata tinggi : {df['tinggi'].mean():.1f} cm")
print(f"Tertinggi      : {df['tinggi'].max()} cm")
print(f"Terendah       : {df['tinggi'].min()} cm")
print(f"Median         : {df['tinggi'].median()} cm")

print("\n" + "="*50)
print("== Calon di atas rata-rata tinggi ==")
rata_tinggi = df['tinggi'].mean()
calon_diatas_rata = df[df['tinggi'] > rata_tinggi][['name', 'alamat', 'tinggi']]
print(calon_diatas_rata.to_string(index=False))

print("\n" + "="*50)
print("== Data dengan Kategori ==")
print(df.to_string(index=False))

print("\n" + "="*50)
print("== Jumlah Calon per Kategori ==")
jumlah_per_kategori = df['kategori'].value_counts()
print(jumlah_per_kategori)

print("\n" + "="*50)
print("== Statistik Lanjutan ==")
print(f"Standar Deviasi : {df['tinggi'].std():.2f}")
print(f"Variansi        : {df['tinggi'].var():.2f}")

print("\n" + "="*50)
print("== Calon dengan Tinggi Badan Tertinggi ==")
tertinggi = df.loc[df['tinggi'].idxmax()]
print(f"Nama   : {tertinggi['name']}")
print(f"Alamat : {tertinggi['alamat']}")
print(f"Tinggi : {tertinggi['tinggi']} cm")
```
